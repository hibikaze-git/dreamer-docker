```python
import pickle

def load_logs(log_path):

  total_logs = []
  with open(log_path, 'rb') as f:
    while True:
      try:
        total_logs.append(pickle.load(f))
      except EOFError:
        break
        
  train_steps = [log['step'] for log in total_logs] 
  train_avg_rews = [log['train_avg_reward'] for log in total_logs]
  eval_steps, eval_avg_rews, eval_std_rews, eval_min_rews, eval_max_rews =[], [], [], [], []
  for log in total_logs:
    try:
      eval_rew = log['eval_avg_reward']
      eval_std = log['eval_std_reward']
      eval_min = log['eval_min_reward']
      eval_max = log['eval_max_reward']
      eval_steps.append(log['step'])
      eval_avg_rews.append(eval_rew)
      eval_std_rews.append(eval_std)
      eval_min_rews.append(eval_min)
      eval_max_rews.append(eval_max)
    except:
      continue 
  return train_steps, train_avg_rews, eval_steps, eval_avg_rews, eval_std_rews, eval_min_rews, eval_max_rews
```

```python
log_path = './data/walker-stand_Dreamerv2_default_hp_09-11-2022-07-17-55/scalar_data.pkl'
train_steps, train_avg_rews, eval_steps, eval_avg_rews, eval_std_rews, eval_min_rews, eval_max_rews = load_logs(log_path)
```

```python
log_path = './data/walker-walk_Dreamerv2_default_hp_08-11-2022-22-52-49/scalar_data.pkl'
train_steps_v2, train_avg_rews_v2, eval_steps_v2, eval_avg_rews_v2, eval_std_rews_v2, eval_min_rews_v2, eval_max_rews_v2 = load_logs(log_path)
```

```python
import matplotlib.pyplot as plt
import os
import numpy as np


def plot_metrics(ax, x, y, stds=None, label=None, linestyle='-'):

    x = np.array(x)
    y = np.array(y)

    # std fill_between

    ax.plot(x, y, label=label, linestyle=linestyle)
    if stds is not None:
        ax.fill_between(x, y-stds, y+stds, alpha=0.2)

    return ax
```

```python
import seaborn as sns

sns.set(style='darkgrid')
fig, ax = plt.subplots(figsize =(10,7))

ax = plot_metrics(ax, x =train_steps[:97], y= train_avg_rews[:97], label='Dreamer')
#ax = plot_metrics(ax, x=train_steps_v2[:97], y=train_avg_rews_v2[:97], label='Dreamerv2')
ax.ticklabel_format(axis='x', style='sci', scilimits=[-5, 3])
ax.set_xlabel('Number of Iterations')
ax.set_ylabel('Rewards')

ax.legend(loc="best")
ax.set_title('cheetah-run training')

save_path = os.path.join("./data/", 'cheetah-run_train.jpg')
fig.savefig(save_path)
```

```python
# eval plot
sns.set(style='darkgrid')
fig, ax = plt.subplots(figsize =(10,7))

ax = plot_metrics(ax, x = eval_steps[:11], y= eval_avg_rews[:11], stds = eval_std_rews[:11], label='Dreamer')
#ax = plot_metrics(ax, x = eval_steps_v2[:11], y= eval_avg_rews_v2[:11], stds = eval_std_rews_v2[:11], label='Dreamerv2')
#ax = plot_metrics(ax, x=eval_steps[:11], y=eval_min_rews[:11], label='min', linestyle='--')
#ax = plot_metrics(ax, x=eval_steps[:11], y=eval_max_rews[:11], label='max', linestyle='--')
ax.ticklabel_format(axis='x', style='sci', scilimits=[-5, 3])
ax.set_xlabel('Number of Iterations')
ax.set_ylabel('Rewards')

ax.legend(loc="best")
ax.set_title('cheetah-run evaluation')

save_path = os.path.join("./data/", 'cheetah-run_eval.jpg')
fig.savefig(save_path)
```
